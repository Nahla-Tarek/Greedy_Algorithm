```python
def type_of_Knapsack():
    x=input("choose the type of knapsack : fractional or  I/0 : ")
    return x
```

```python
def cost_function():
    x=input("choose the cost function : Maximum Profit - Minimum Weight - Maximum Profit/Weight : ")
    return x
   
```

```python
import pandas as pd
import numpy as np
from copy import deepcopy
def greedy(data,capacity):
    df = pd.read_csv(data)
    print("Dataset :")
    print(df)
    print(" ")
    r=type_of_Knapsack()
    print(" ")
    e=cost_function()
    if e=="Maximum Profit" :
        df1=df.sort_values("profit", ascending=False)
        print(" ")
        print('This is the data set that is sorted according to the profit column descending :')
        print(df1)
        df2= deepcopy(df1)
        df2["weight"]=df2["weight"].cumsum() ## cumulative sum for the weight column
        print(" ")
        print("Data set after getting cumulative sum for the weight column : ")
        print(df2)
        
      
    elif e=="Minimum Weight":
        df1=df.sort_values("weight", ascending=True)
        print(" ")
        print('This is the data set that is sorted according to the weight column ascending :')
        print(df1)
        df2= deepcopy(df1)
        df2["weight"]=df2["weight"].cumsum()  ## cumulative sum for the weight column
        print(" ")
        print("Data set after getting cumulative sum for the weight column : ")
        print(df2)
        
    else:
        df1=df.sort_values("profit/weight", ascending=False)
        print(" ")
        print('This is the data set that is sorted according to the profit/weight column descending :')
        print(df1)
        df2= deepcopy(df1)
        df2["weight"]=df2["weight"].cumsum()  ## cumulative sum for the weight column
        print(" ")
        print("Data set after getting cumulative sum for the weight column : ")
        print(df2)
        
       
    if r=="fractional":   ## fractional knapstack
        df2["weight1"]=df1["weight"]
        df2["greed"]  = np.where(df2['weight'] < capacity,0,(df2['weight']-capacity)) ## comparing the capacity with objects weight
        df2["new_weight"]=df2["weight1"]-df2["greed"] ## get the new weight (fraction) of last object can be added to the knapsack
        df3=df2.cumsum()
        df2["new_weight1"]=df3["new_weight"]
        select_row = df2.loc[df2["greed"]==0]
        select_row1 = df2.loc[(df2["new_weight1"]==capacity) & (df2["new_weight"]!=0)&(df2["greed"]!=0)]
        select_row = select_row.append(select_row1)
        df4=select_row[["object","new_weight"]]
        print(" ")
        print("dataset with the objects that can be added to knapsack : ")
        print(select_row)
        print(" ")
        print(" new data set with two columns to create a dictionary acts as knapsack : ")
        print(df4)
        select_row["max_profit"]=select_row ["profit/weight"]*select_row ["new_weight"]
        max_profit=select_row["max_profit"].sum()
        print(" ")
        print("knapsack contains : ", df4.to_dict('records'))
        print(" ")
        print("maximum profit is : ", max_profit)
           
    else:           ## 1-0 knapstack 
        df2["weight1"]=df1["weight"]
        df2["greed"]  = np.where(df2['weight'] < capacity,0,(df2['weight']-capacity)) ## comparing the capacity with objects weight
        df2["new_weight"]=df2["weight1"]-df2["greed"]
        sel=df2.loc[df2["greed"]!=0]
        select_row = df2.loc[df2["greed"]==0]
        a=list(sel["weight1"])
        c=select_row["new_weight"].sum()
        for i in a:
            b=i+c
            if b<=15:
                select_row = select_row.append(df2.loc[(df2["weight1"]==i)&(df2["greed"]!=0)])
                c=b
        df4=select_row[["object","weight1"]]
        print(" ")
        print(" dataset with the objects that can be added to knapsack: ")
        print(select_row)
        print(" ")
        print(" new data set with two columns to create a dictionary acts as knapsack : ")
        print(df4)
        max_profit=select_row["profit"].sum()
        print(" ")
        print("knapsack contains : ", df4.to_dict('records'))
        print(" ")
        print("maximum profit is : ", max_profit)
       
   
        
greedy('greedy2.csv',50)
```
